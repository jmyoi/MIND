```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import os

import seaborn as sns

CKEYS = plt.rcParams['axes.prop_cycle'].by_key()['color']
font = {'family' : 'arial',
        'weight' : 'regular',
        'size'   : 15}
import matplotlib
matplotlib.rc('font', **font)
```

```python
df_meteorite = pd.read_csv("Meteorite_Landings.csv")
```

```python
df_meteorite = df_meteorite.dropna()
```

```python
df_meteorite['year'] = df_meteorite['year'].str.slice(6, 10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_meteorite['year'], df_meteorite['mass (g)'],'.')
_ = plt.xticks(visible=False)
```

```python
df_meteorite.head()
```

```python
color_yr = [int(year)/2018. for year in df_meteorite['year']]
```

```python
plt.figure(figsize=(15,8))
plt.scatter(df_meteorite['reclong'], df_meteorite['reclat'],
            c=color_yr, cmap=plt.cm.autumn)
plt.colorbar()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

```python
fg = sns.FacetGrid(data=df_meteorite, row='reclong', col='reclat', hue='recclass')
fg.map(plt.scatter, 'Longitude', 'Latitude').add_legend()
```
